### **Option 1:** Manually get SMILES for the given Pubchem CID with Activity outcome

This is needed for the AID 1706 dataset. It has around 300000 compounds. while downloading the SMILES it does not work for huge amount of compounds at one go. So, we need to manually set the value in the input file from where we have to start downloading the SMILES. 

```python
data_1 = pd.read_csv("../data/pubchem_data/AID_1706_datatable_all.csv")
data_1
```

```python
from pubchempy import *

processed_df = pd.DataFrame(columns = ['Smiles', 'PUBCHEM_CID', 'PUBCHEM_ACTIVITY_OUTCOME'])

start_idx = 160002

for i in range(processed_df.shape[0]+start_idx,data_1.shape[0]):
  cId = data_1.iloc[i,2]
  c = Compound.from_cid(str(cId))
  cls = data_1.iloc[i,3]
  sm = c.canonical_smiles
  #print(i, '\t', sm, '\t', cId, cls)
  print(i, ",", end=" ")
  s1 =  pd.DataFrame([[sm, cId, cls]], columns=['Smiles', 'PUBCHEM_CID', 'PUBCHEM_ACTIVITY_OUTCOME']  )
  processed_df = pd.concat([processed_df, s1], ignore_index=True)
```

```python
filename = '../data/processed/'+"AID_1706_datatable_all"+"_smiles_"+str(start_idx)+".csv"
print(filename)
processed_df.to_csv(filename, index=False) 
```

### **Option 2:** Function for Pubchem CID to SMILES

Given the database of pubchem that has the pubchem compound id,extract the canonical SMILEs in a dataframe.

- *Input:* filepath that contains the Pubchem Compound ID and the Activity Outcome
- *Output:* dataframe that contains the Canonical SMILES, CID and Activity Outcome

```python
from pubchempy import *
import pandas as pd

def create_smiles(path_sample_details_1):
  data_1 = pd.read_csv(path_sample_details_1)
  data_0 = pd.DataFrame()
  data_0['PUBCHEM_CID'] = data_1['PUBCHEM_CID']
  data_0['PUBCHEM_ACTIVITY_OUTCOME'] = data_1['PUBCHEM_ACTIVITY_OUTCOME']
  n = 0
  smiles = []
  for cid in data_0['PUBCHEM_CID']:
    c = Compound.from_cid(cid)
    smiles.append(c.canonical_smiles)
    n = n +1
    #print(c.canonical_smiles, n)
  print("Total Data Processed: ", n)
  data_0['Smiles'] = smiles

  data_0 = data_0[['Smiles', 'PUBCHEM_CID', 'PUBCHEM_ACTIVITY_OUTCOME']]
  fileName = path_sample_details_1.split('/')[-1].split('.')[0]
  filename = '../data/processed/'+fileName+"_smiles.csv"
  print(filename)
  data_0.to_csv(filename, index=False)
  return data_0

```
